# Logistic Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('./train.csv')
dataset = dataset.dropna(how='any', subset=['Embarked'])
X_train = dataset.iloc[:, [2, 4, 6, 7, 9, 11]].values
y_train = dataset.iloc[:, 1].values

dataset = pd.read_csv('./test.csv')
X_test = dataset.iloc[:, [1, 3, 5, 6, 8, 10]].values
y_pid = dataset.iloc[:, [0]].values
```

## Printing data

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

## Label encode M/F

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_train[:,1] = le.fit_transform(X_train[:,1])
print(X_train)

X_test[:,1] = le.transform(X_test[:,1])
print(X_test)
```

## Onehot encode station

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [5])], remainder='passthrough') # Transforms last column
X_train = np.array(ct.fit_transform(X_train))

X_test = np.array(ct.transform(X_test))
```

```python
np.set_printoptions(threshold=np.inf)
print(X_train)
```

```python
print(X_test)
```

## Take care of missing fare in test data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X_train[:, 7:8])
X_test[:, 7:8] = imputer.transform(X_test[:, 7:8])

#X_test[-1].fillna(X_train[-1].mean(), inplace=True)

print(X_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) # Scaling all IVs
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the Logistic Regression model on the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
output=np.concatenate((y_pid.reshape(len(y_pid),1), y_pred.reshape(len(y_pred),1)),1)
print(output)
np.savetxt("result_linear_regression.csv", output, delimiter=',', header='PassengerId,Survived', fmt='%d', comments='')
```

##  Training the Kernel SVM model on the Training set

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
output=np.concatenate((y_pid.reshape(len(y_pid),1), y_pred.reshape(len(y_pred),1)),1)
print(output)
np.savetxt("result_kernel_svm.csv", output, delimiter=',', header='PassengerId,Survived', fmt='%d', comments='')
```

##  Training the Random Forest model on the Training set

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
output=np.concatenate((y_pid.reshape(len(y_pid),1), y_pred.reshape(len(y_pred),1)),1)
print(output)
np.savetxt("result_random_forest.csv", output, delimiter=',', header='PassengerId,Survived', fmt='%d', comments='')
```

##  Training the xGBoost model on the Training set

```python
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
output=np.concatenate((y_pid.reshape(len(y_pid),1), y_pred.reshape(len(y_pred),1)),1)
print(output)
np.savetxt("result_xgboost.csv", output, delimiter=',', header='PassengerId,Survived', fmt='%d', comments='')
```
